```python
import time
import random
import matplotlib.pyplot as plt
import heapq


def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key


def heap_sort(arr):
    heapq.heapify(arr)
    sorted_arr = []
    while arr:
        sorted_arr.append(heapq.heappop(arr))
    return sorted_arr


def time_insertion_sort(n):
    arr = [random.randint(1, 10000) for _ in range(n)]
    start_time = time.time()
    insertion_sort(arr)
    return time.time() - start_time


def time_heap_sort(n):
    arr = [random.randint(1, 10000) for _ in range(n)]
    start_time = time.time()
    heap_sort(arr)
    return time.time() - start_time

n_values = []
times_insertion_sort = []
times_heap_sort = []

for n in range(1000, 10001, 1000):
    n_values.append(n)
    times_insertion_sort.append(time_insertion_sort(n))
    times_heap_sort.append(time_heap_sort(n))


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.plot(n_values, times_insertion_sort, label="Insertion Sort", marker='o', color='blue')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Insertion Sort: Time vs n')
plt.legend()
plt.grid(True)


plt.subplot(1, 2, 2)
plt.plot(n_values, times_heap_sort, label="Heap Sort", marker='o', color='red')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Heap Sort: Time vs n')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
